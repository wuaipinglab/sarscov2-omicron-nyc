## Figure 1B: Temporal distribution of 392 Omicron viruses in New York City.

```python
import os
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore')
```

```python
meta = pd.read_csv('../data/rawdata/nyc_omicron_metadata.csv', index_col='seqName')
```

```python
nextclade = pd.read_csv('../data/nextclade_nyc/nextclade.tsv', delimiter='\t', index_col='seqName')
```

```python
dstart = meta['Date of Collection'].sort_values().values[0]
dend = meta['Date of Collection'].sort_values().values[-1]
dates = [str(i)[:10] for i in pd.date_range(dstart, dend)]

counts = []
d_counts = meta['Date of Collection'].value_counts()
for d in dates:
    if d in d_counts:
        counts.append(d_counts[d])
    else:
        counts.append(0)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 10), dpi=300)

plt.bar(dates, counts, edgecolor='black', color='#0274B0', alpha=0.7)

ax.spines['bottom'].set_linewidth(3)
ax.spines['left'].set_linewidth(3)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.xticks(dates, ['Nov-25', '', '', '', 'Nov-29', '', '', '', 'Dec-03', '', '', '', 'Dec-07', '', '', '', 'Dec-11'])

plt.xlim(-1, 17)
plt.ylim(0, 132)

plt.savefig('../figures/Figure_1B.pdf')
```
